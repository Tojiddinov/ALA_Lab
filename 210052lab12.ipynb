```python
import numpy as np
import matplotlib.pyplot as plt

 #Problem 1
y = np.linspace(-1,1,20)
orig = np.vstack((np.zeros(y.shape),y))

th = np.pi/3
T = np.array([[np.cos(th), np.sin(th)],[-np.sin(th), np.cos(th)]])
t = T@orig

plt.figure(figsize = (5,5))
plt.plot(orig[0,:],orig[1,:], 'ko' , label = "Original")
plt.plot(t[0,:],t[1,:],'s',color = "gray" , label = "Transformed")
plt.xlim(-1.2,1.2)
plt.ylim(-1.2,1.2)
plt.legend()
plt.show()

 #Problem 2
y = np.linspace(-1,1,20)
orig = np.vstack((np.zeros(y.shape),y))

fig,ax= plt.subplots(1,figsize = (5,5))
ax.set_xlim([-1.5,1.5])
ax.set_ylim([-1.5,1.5])
ax.plot(orig[0,:],orig[1,:],"ko")
plth, = ax.plot(orig[0,:], orig[1,:], "ko")
phi = np.linspace(0,2*np.pi,20)

import matplotlib.animation as animation
from matplotlib import rc
rc('animation',html = 'jshtml')

def aframe(phi):
  T = np.array([[np.cos(phi), np.sin(phi)],[-np.sin(phi), np.cos(phi)]])
  t = T@orig

  plth.set_xdata(t[0,:])
  plth.set_ydata(t[1,:])
  return plth

animation.FuncAnimation(fig,aframe,phi,interval = 100, repeat = True)
```

```python
#problem3
th = np.linspace(0,2*np.pi,20)
orig = np.vstack((np.cos(th), np.sin(th)))
T = np.array([[1, 0.5],[0,0.5]])
t = T@orig

plt.figure(figsize = (5,5))
plt.plot(orig[0,:],orig[1,:], 'ko' , label = "Original")
plt.plot(t[0,:],t[1,:],'s',color = "gray" , label = "Transformed")
plt.xlim(-2,2)
plt.ylim(-2,2)
plt.legend()
plt.show()
```

```python
#problem4
th = np.linspace(0,2*np.pi,20)
orig1 = np.vstack((th, np.cos(th)))
orig2 = np.vstack((th, np.sin(th)))

fig,ax= plt.subplots(1,figsize = (10,4))
ax.set_xlim([-2,4])
ax.set_ylim([-2,2])

plth1, = ax.plot(orig1[0,:], orig1[1,:], "ko")
plth2, = ax.plot(orig2[0,:], orig2[1,:], 's',color = "gray")

phi = 1 - np.linspace(-1,1-1/40,40)

def aframe(phi):
  T = np.array([[(1-phi)/3, 0],[0, phi]])
  t1 = T@orig1
  t2 = T@orig2
  plth1.set_xdata(t1[0,:])
  plth1.set_ydata(t1[1,:])
  plth2.set_xdata(t2[0,:])
  plth2.set_ydata(t2[1,:])
  return plth1,plth2

animation.FuncAnimation(fig,aframe,phi,interval = 100, repeat = True)
```
