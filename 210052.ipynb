```python
import numpy as np
a1=3.1
a2=4.9
a3=5.1


v1=np.array([6, 2, 0, 4])
v2=np.array([-1, 5, 4, 9])
v3=np.array([2, 1, 10, 1])
a1*v1 + a2*v2 + a3*v3
```

```python
a = np.array([a1, a2, a3])
v = np.array([v1, v2, v3])
result = np.zeros(len(v1))
for s, vv in zip(a, v):
    result+=s*vv
result
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

v1 =np.array([6, 1, 4]) 
v2 =np.array([2, 0, -3])
fig = plt.figure()
ax=fig.add_subplot(111, projection='3d')
ax.quiver(0, 0, 0, v1[0], v1[1], v1[1], v1[2], color = 'r', arrow_length_ratio = 0.5)
ax.quiver(0, 0, 0, v2[0], v1[1], v1[1], v1[2], color = 'b', arrow_length_ratio = 0.5)
ax.set_xlim([-8, 8])
ax.set_ylim([-8, 8])
ax.set_zlim([-8, 8])
plt.show()
```
